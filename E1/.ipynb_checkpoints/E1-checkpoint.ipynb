```python
import csv
import matplotlib.pyplot as plt
import numpy as np
import itertools as it
```

**En esta parte del notebook, se da inicio a la primera csv que seria "pacientesvmi"; Dentro del codigo se comento al final una linea la cual se uso para comprobar con menos datos en el grafico para que se pudiera entender**

```python
data_vmi={}
casos=[]
pacientes_vmi=[]
pacientes_no_vmi=[]
pacientes_vmi_noi=[]
with open('data/PacientesVMI.csv') as archivo_csv:
    lector = csv.reader(archivo_csv, delimiter=',')
    contador = 0
    for row in lector:
        if contador == 0:
            casos = row
            contador=1
            casos.remove("Casos")
        elif contador == 1:
            pacientes_vmi = row
            contador=2
            pacientes_vmi.remove("Pacientes VMI")
        elif contador == 2:
            pacientes_no_vmi = row
            contador=3
            pacientes_no_vmi.remove('Pacientes no VM')
        elif contador == 3:
            pacientes_vmi_noi = row
            pacientes_vmi_noi.remove('Pacientes VM no invasiva')
    for idx in range(len(casos)):
            data_vmi[casos[idx]] = dict(pacientes_vmi=float(pacientes_vmi[idx] or 0), pacientes_no_vmi=float(pacientes_no_vmi[idx] or 0), pacientes_vmi_noi=float(pacientes_vmi_noi[idx] or 0))
    #data_vmi=dict(it.islice(data_vmi.items(),200,210))
```

**Aca se comprueba que el codigo de arriba funcione correctamente y muestre los datos de una manera "ordenada"**

```python
data_vmi
```

**Graficar los datos de la primera csv**

```python
fig,ax = plt.subplots(figsize=(30,20))

etiquetas=list(data_vmi.keys())

pacientes_vmi=[]
pacientes_no_vmi=[]
pacientes_vmi_noi=[]
for etiqueta in etiquetas:
    pacientes_vmi.append(data_vmi[etiqueta]['pacientes_vmi'])
    pacientes_no_vmi.append(data_vmi[etiqueta]['pacientes_no_vmi'])
    pacientes_vmi_noi.append(data_vmi[etiqueta]['pacientes_vmi_noi'])

ejeX = np.arange(len(etiquetas))

ancho=0.5

barraR = ax.bar(ejeX - ancho/3, height=pacientes_vmi, label='pacientes_vmi',color="lightgreen", width=ancho/3, align='center')
barraA = ax.bar(ejeX , height=pacientes_no_vmi, label='pacientes_no_vmi',color="red", width=ancho/3, align='center')
barraB = ax.bar(ejeX + ancho/3, height=pacientes_vmi_noi, label='pacientes_vmi_noi',color="blue", width=ancho/3, align='center')

ax.set_ylabel('Cantidad de pacientes con ventilación mecánica')

ax.set_title('Informe de covid')

ax.set_xticks(ejeX)

ax.set_xticklabels(etiquetas, rotation=90)

ax.legend()
```

**Se da comienzo al segundo csv del ejercicio; Se dejo codigo comentado para limitar los casos y asi dar una imagen mas comprensible a la hora de graficar**

```python
data={}
with open('data/activos_vs_recuperados.csv') as archivo_csv:
    lector = csv.reader(archivo_csv, delimiter=',')
    for i, row in enumerate(lector):
        if row[1]!='activos':
            data[row[0]] = dict(activos=int(row[1]), recuperados=int(row[2]))
            #if i==20:
                #break
```

**se vuelve a comprobar que los datos esten ordenados**

```python
data
```

**Se procede a graficar**

```python
fig,ax = plt.subplots(figsize=(30,20))

etiquetas=list(data.keys())

activos=[]
recuperados=[]
for etiqueta in etiquetas:
    activos.append(data[etiqueta]['activos'])
    recuperados.append(data[etiqueta]['recuperados'])

ejeX = np.arange(len(etiquetas))

ancho=0.5

barraR = ax.bar(ejeX - ancho/2, recuperados, ancho, label='recuperados',color="lightgreen")
barraA = ax.bar(ejeX + ancho/2, activos, ancho, label='activos',color="red")

ax.set_ylabel('Cantidad de pacientes activos v/s recuperados')

ax.set_title('Informe de covid')

ax.set_xticks(ejeX)

ax.set_xticklabels(etiquetas, rotation=90)

ax.legend()
```
